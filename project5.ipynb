```python
# How many people purchase during AM
# Use count() to count number of records
df[df['AM or PM'] == 'AM']['Email'].count()
```

```python
# How many people purchase during PM
df[df['AM or PM'] == 'PM']['Email'].count()
```

```python
# How many people have a Credit Card that expires in 2020
df[df['CC Exp Date'].str.contains('/20')]['Email'].count()
# Count number of values in CC Exp Date column that contains '/20' which represents year 2020
# Other way to do it is to create user function and iterate with for loop to count number of values
```

```python
# What are the top 5 most popular email providers/hosts (e.g. gmail.com, yahoo.com, etc...)

df['Email'].str.split('@').str[1].value_counts().head(5)
# Grab the email column and split it into two columns, one with the username and one with the host. Result is a list of tuples. Take the 2nd element (host) since the first is the username.
# Then, use value_counts() to count the number of times each host was used
# Default sorting order is descending, so we use head(5) to get the top 5
```
